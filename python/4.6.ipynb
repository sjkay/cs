```python
from skimage import novice,data
class MyPic():
    def __init__(self,filename):
        self.pic = novice.open(data.data_dir + "/" + filename)
    def threshold(self,p):
        if p > 255:
            return 255
        else:
            return int(p)
    def show(self):
        self.pic.show()
    def sum(self):
        (r,g,b,a) = (0,0,0,0)
        for p in self.pic:
            r += p.red
            g += p.green
            b += p.blue
            a += p.alpha
        return (r,g,b,a)
    def grey(self):
        for p in self.pic:
            (r,g,b) = (p.red, p.green, p.blue)
            grey = self.threshold((r + g + b)/3)
            (p.red,p.green,p.blue)=(grey,grey,grey)
```

```python
pic = MyPic("chelsea.png")
pic.show()
```

```python
pic = MyPic("chelsea.png")
print(pic.sum())
```

```python
pic = MyPic("chelsea.png")
print(sum(pic.sum()))
```

```python
pic = MyPic("chelsea.png")
pic.grey()
pic.show()
```
