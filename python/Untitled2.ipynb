```python
from skimage import novice
from skimage import data
class Mypic:
    def read(self, name):
        self.pic = novice.open(data.data_dir + "/" + name)
        return self.pic
    def paris(self):
        pic = self.pic
        for y in range(pic.height):
            for x in range(pic.width//3):
                pic[x,y].blue = 255
                pic[x+pic.width//3,y].alpha = 120
                pic[x+2*pic.width//3,y].red = 255
        return pic
pic = Mypic()
pic.read('coffee.png')
pic.paris()
```

```python
def palindrome(s):
    for i in range(len(s)):
        if s[i] != s[len(s)-1-i]:
            return False
    return True
```

```python
palindrome('amadama')
```

```python
def mysplit(s):
    lst = []
    tmp = ''
    for c in s:
        if c != ' ':
            tmp = tmp + c
        else:
            lst.append(tmp)
            tmp = ''
    if tmp != '':
        lst.append(tmp)
        tmp = ''
    return lst
```

```python
s = 'This is right justified'
print(mysplit(s))
```
