```python
import time
def is_prime(n):
    if n > 1:
        for i in range(2,n):
            if n%i == 0:
                return False
        return True
    else:
        return False
def all_primes(n):
    lst = []
    for i in range(1,n+1):
        if is_prime(i):
            lst.append(i)
    return lst
```

```python
cur = time.time()
print(len(all_primes(10000)))
print(time.time() - cur)
```

```python
cur = time.time()
print(len(all_primes(100000)))
print(time.time() - cur)
```

```python
import time
def sieve(n):
    lst = [i for i in range(2,n+1)]
    for k in lst:
        if k > 0:
            for i in range(2,n):
                p = k * i
                if p <= n:
                    lst[p - 2] = 0
                else:
                    break
    return [x for x in lst if x > 0]
```

```python
cur = time.time()
print(len(sieve(10000)))
print(time.time() - cur)
```

```python
from math import sqrt
def mrange(start, stop, step):
    while start < stop:
        yield start
        start += step
def is_prime(num):
    if num == 2:
        return True
    if (num < 2) or (num % 2 == 0):
        return False
    return all(num % i for i in mrange(3, int(sqrt(num)) + 1, 2))
```

```python
def check_prime_lst(lst):
    for x in lst:
        if not is_prime(x):
            return False
    return True
print(check_prime_lst(lst))
```
